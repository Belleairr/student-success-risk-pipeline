```python
import pandas as pd
import sqlite3

# Step 1: Loading the Student Success dataset
# Using sep=';' because the file uses semicolons
df = pd.read_csv('data/raw/data.csv', sep=';')

# Step 2: Basic logical cleaning using Pandas (easier than SQL for names with spaces)
# We keep rows where Age is > 0 and 1st sem grades are >= 0
df_clean = df[(df['Age at enrollment'] > 0) & (df['Curricular units 1st sem (grade)'] >= 0)].copy()

# Step 3: Standardizing the column names for the team
# We'll replace spaces with underscores so Ellen doesn't have issues later
df_clean.columns = [c.replace(' ', '_').replace('(', '').replace(')', '') for c in df_clean.columns]

print(f"Success! Total students after cleaning: {len(df_clean)}")
df_clean.head()
```

```python
# Step 4: Saving the final version to our project database
conn = sqlite3.connect('student_data.db')
df_clean.to_sql('cleaned_students', conn, if_exists='replace', index=False)

# Step 5: Exporting the 'Gold Standard' CSV for the team
df_clean.to_csv('data/processed/cleaned_student_data.csv', index=False)

# Closing the connection
conn.close()

print(" Data successfully exported to SQL and the processed folder!")
```
