```python
import pandas as pd

# 1. Load the data
# 'python' to help prevent the notebook from hanging
df = pd.read_csv('data/raw/data.csv', sep=';', engine='python')

# 2. Clean the headers (Adding .strip() to remove that \t character)
df.columns = [c.strip().replace(' ', '_').replace("'", "").replace('(', '').replace(')', '').replace('/', '_').lower() for c in df.columns]

# 3. Filter for valid data
df_clean = df[(df['age_at_enrollment'] > 0) & (df['curricular_units_1st_sem_grade'] >= 0)].copy()

# 4. Save to the processed folder
df_clean.to_csv('data/processed/cleaned_student_data.csv', index=False)

print(f"✅ Success! Cleaned {len(df_clean)} records using only Pandas.")
df_clean.head()
```

```python
# Convert target to binary classification 
# Dropout = 1, everything else = 0
df_clean['target'] = df_clean['target'].apply(lambda x: 1 if x == 'Dropout' else 0)

# see the new counts
print("Target Variable Distribution (1 = Dropout, 0 = Not Dropout):")
print(df_clean['target'].value_counts())


df_clean.to_csv('data/processed/cleaned_student_data.csv', index=False)
```
