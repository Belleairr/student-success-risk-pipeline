To Do:
- look into simplifying Parents Occupation and their Qualifications
- Value 99 is blank for both Parents Occupations
- Drop any features that risk data leakage

```python
#simplify/replace parents occupation

def simplify_occupation(x):
    if x in [1, 2, 122, 123, 125]:
        return "Highly Skilled"
    elif x in [3, 7, 131]:
        return "Skilled"
    elif x in [4, 5, 141, 143]:
        return "Administrative"
    elif x in [9, 191, 192, 193]:
        return "Unskilled"
    elif x in [10, 101, 102, 103]:
        return "Military"
    else:
        return "Other"

cur_df['mothers_occupation'] = cur_df['mothers_occupation'].apply(simplify_occupation)
cur_df['fathers_occupation'] = cur_df['fathers_occupation'].apply(simplify_occupation)
```

```python
#simplify/replace parents qualifications

def simplify_education(x):
    if x in [35, 36]:
        return "No Education"
    elif x in [9, 10, 11, 26, 29]:
        return "Basic"
    elif x in [1, 12, 14, 19]:
        return "Secondary"
    elif x in [2, 3, 4, 5, 40, 43, 44]:
        return "Higher"
    else:
        return "Other"

cur_df['mothers_qualification'] = cur_df['mothers_qualification'].apply(simplify_education)
cur_df['fathers_qualification'] = cur_df['fathers_qualification'].apply(simplify_education)
```

```python
#simplify courses feature

def simplify_courses(x):
    if x in [33]:
        return "Engineering/Technology"
    elif x in [9119]:
        return "Computer Science/IT"
    elif x in [9003, 9130]:
        return "Agriculture/Animal Sciences"
    elif x in [9085, 9500, 9556]:
        return "Health"
    elif x in [9147, 9991]:
        return "Business/Management"
    elif x in [9254, 9670]:
        return "Business/Services"
    elif x in [8014, 9238]:
        return "Social Services"
    elif x in [9853]:
        return "Education"
    elif x in [171, 9070]:
        return "Design/Creative"
    elif x in [9773]:
        return "Media/Communication"
    else:
        return "Other"

cur_df['course'] = cur_df['course'].apply(simplify_courses)
```

```python
#One Hot Encode Parents qualifications + occupations + courses

cols_to_encode = [
    'mothers_occupation',
    'fathers_occupation',
    'mothers_qualification',
    'fathers_qualification',
    'course'
]

cur_df = pd.get_dummies(cur_df, columns=cols_to_encode, drop_first=True)
```

To Do:
- Simplify Application_mode
- Investigate class imbalance and next steps

```python
def simplify_application_mode(x):
    if x in [1, 17, 18]:
        return "Regular Admission"

    elif x in [42, 43, 51, 57]:
        return "Transfer/Change"

    elif x in [5, 16]:
        return "Special Entry"

    elif x in [7, 15, 39, 44, 53]:
        return "Non-Traditional"

    else:
        return "Other"

cur_df['application_mode'] = cur_df['application_mode'].apply(simplify_application_mode)
cur_df = pd.get_dummies(cur_df, columns=['application_mode'], drop_first=True)
```

```python
cur_df.head(5)
```

```python
#ensure the one hot encoded features are 0s and 1s and not True/False

bool_cols = cur_df.select_dtypes(include='bool').columns
cur_df[bool_cols] = cur_df[bool_cols].astype(int)
```
